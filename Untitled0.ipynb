```python
import pandas as pd
import numpy as np

# Convert time to datetime
depth_df['Time'] = pd.to_datetime(depth_df['Time'].str.replace(' +0530 IST', ''), utc=True)
volume_df['Time'] = pd.to_datetime(volume_df['Time'].str.replace(' +0530 IST', ''), utc=True)

# Sort both (just in case)
depth_df.sort_values("Time", inplace=True)
volume_df.sort_values("Time", inplace=True)

# --- Midprice, Spread, Microprice, Imbalance ---

# Top of book
depth_df['bid_1_price'] = depth_df['BidPriceL20']   # Best bid
depth_df['ask_1_price'] = depth_df['AskPriceL20']   # Best ask
depth_df['bid_1_qty']   = depth_df['BidQtyL20']
depth_df['ask_1_qty']   = depth_df['AskQtyL20']

depth_df['mid_price'] = (depth_df['bid_1_price'] + depth_df['ask_1_price']) / 2
depth_df['spread'] = depth_df['ask_1_price'] - depth_df['bid_1_price']
depth_df['microprice'] = (depth_df['bid_1_price'] * depth_df['ask_1_qty'] +
                          depth_df['ask_1_price'] * depth_df['bid_1_qty']) / \
                         (depth_df['bid_1_qty'] + depth_df['ask_1_qty'])

depth_df['imbalance_lvl1'] = (depth_df['bid_1_qty'] - depth_df['ask_1_qty']) / \
                             (depth_df['bid_1_qty'] + depth_df['ask_1_qty'])

# --- Rolling Volatility ---
depth_df.set_index('Time', inplace=True)
depth_df['log_return'] = np.log(depth_df['mid_price'] / depth_df['mid_price'].shift(1))
depth_df['volatility_10s'] = depth_df['log_return'].rolling('10s').std()
depth_df['volatility_30s'] = depth_df['log_return'].rolling('30s').std()

# --- Cumulative Depth ---
bid_qty_cols = [col for col in depth_df.columns if 'BidQtyL' in col]
ask_qty_cols = [col for col in depth_df.columns if 'AskQtyL' in col]

depth_df['cum_bid_qty'] = depth_df[bid_qty_cols].sum(axis=1)
depth_df['cum_ask_qty'] = depth_df[ask_qty_cols].sum(axis=1)

# Reset index for merging later
depth_df.reset_index(inplace=True)

# --- Volume Features from volume_df ---

# Define buy vs sell using IsMarketMaker
volume_df['side'] = np.where(volume_df['IsMarketMaker'], 'sell', 'buy')

# Resample per second
volume_df.set_index('Time', inplace=True)
volume_1s = volume_df.groupby('side').resample('1s')['Quantity'].sum().unstack(0).fillna(0)
volume_1s['vol_imbalance'] = (volume_1s['buy'] - volume_1s['sell']) / (volume_1s['buy'] + volume_1s['sell'] + 1e-6)

# Rolling VWAP (volume-weighted average price)
volume_df['vwap'] = volume_df['Price'] * volume_df['Quantity']
vwap_rolling = volume_df['vwap'].resample('1s').sum() / volume_df['Quantity'].resample('1s').sum()
vwap_rolling = vwap_rolling.fillna(method='ffill')

# Merge with depth features
final_df = pd.merge_asof(depth_df.sort_values('Time'),
                         volume_1s.reset_index().sort_values('Time'),
                         on='Time')

final_df['vwap'] = vwap_rolling.reset_index(drop=True)

# VWAP shift over time
final_df['vwap_shift_10s'] = final_df['vwap'].diff(10)
final_df['vwap_shift_30s'] = final_df['vwap'].diff(30)

print("✅ Feature engineering complete. Final columns:", final_df.columns.tolist())
```

```python
engineered_cols = [
    'mid_price', 'spread', 'microprice', 'imbalance_lvl1',
    'log_return', 'volatility_10s', 'volatility_30s',
    'cum_bid_qty', 'cum_ask_qty', 'buy', 'sell',
    'vol_imbalance', 'vwap', 'vwap_shift_10s', 'vwap_shift_30s'
]

X = final_df[engineered_cols].dropna().copy()


```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# Normalize
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Clustering - you can switch to DBSCAN/HDBSCAN later
kmeans = KMeans(n_clusters=4, random_state=42)
clusters = kmeans.fit_predict(X_pca)

# Add to DataFrame
X['pca_1'] = X_pca[:, 0]
X['pca_2'] = X_pca[:, 1]
X['cluster'] = clusters

# Visualize
plt.figure(figsize=(10, 6))
sns.scatterplot(data=X, x='pca_1', y='pca_2', hue='cluster', palette='Set2', s=40)
plt.title("Regime Clusters via PCA + KMeans")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.grid(True)
plt.legend(title="Regime")
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Select engineered features
engineered_cols = [
    'mid_price', 'spread', 'microprice', 'imbalance_lvl1',
    'log_return', 'volatility_10s', 'volatility_30s',
    'cum_bid_qty', 'cum_ask_qty', 'buy', 'sell',
    'vol_imbalance', 'vwap', 'vwap_shift_10s', 'vwap_shift_30s'
]

X = final_df[engineered_cols].dropna().copy()

# Step 2: Normalize
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 3: PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Step 4: Clustering (KMeans)
kmeans = KMeans(n_clusters=4, random_state=42)
clusters = kmeans.fit_predict(X_pca)

# Step 5: Add to DataFrame
X['pca_1'] = X_pca[:, 0]
X['pca_2'] = X_pca[:, 1]
X['cluster'] = clusters

# Step 6: Visualize Clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(data=X, x='pca_1', y='pca_2', hue='cluster', palette='Set2', s=50)
plt.title("🧠 Market Regimes - KMeans Clustering (PCA 2D)")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.grid(True)
plt.legend(title="Regime Cluster")
plt.show()
```

```python
import hdbscan

# Step 1: HDBSCAN Clustering
clusterer = hdbscan.HDBSCAN(min_cluster_size=50, metric='euclidean')
hdb_clusters = clusterer.fit_predict(X_scaled)

# Step 2: Add to DataFrame
X['hdb_cluster'] = hdb_clusters

# Step 3: Visualize HDBSCAN Clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(data=X, x='pca_1', y='pca_2', hue='hdb_cluster', palette='tab10', s=50)
plt.title("🔍 HDBSCAN Market Regimes (PCA 2D)")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.grid(True)
plt.legend(title="HDB Regime")
plt.show()
```
